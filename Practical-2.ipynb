```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
objects = ('python','c++','java','perl','scala','lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
plt.bar(y_pos,performance,width=0.5, align='center',alpha=1, color=['r','r','g','g','b','b'])
plt.xticks(y_pos,objects)
plt.ylabel('Usage')
plt.xlabel('Programming language')
plt.title('Programming language usage')
```

```python
import matplotlib.pyplot as plt 
country = ['brazi','usa','russia','spain','uk','india']
totalDeaths = [112596,37312,5971,27136,40597,7449]

plt.bar(country,totalDeaths,width=0.9,align='center',color='cyan',edgecolor='red')

i=1.0
j=2000

for i in range(len(country)):
    plt.annotate(totalDeaths[i],(-0.3+i,totalDeaths[i]+j))
    
plt.legend(labels=['Total Deaths'])
plt.title("Bar plot representing the total deaths by top 6 countries due to coronavirus")
plt.xlabel('Country')
plt.ylabel('deaths')
plt.savefig('1BarPlot.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

Data = {'Tasks':[300,500,700],
       'Task Type':['Task Pending','Task Ongoing','Task Completed']}
df = pd.DataFrame(Data)
df.set_index('Task Type',inplace=True)
df.plot.pie(y = 'Tasks', figsize=(5,5),autopct='%1.1f%%',startangle=90)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

label=['civil','electrical','Mechanical','chemicals']
sizes=[15,30,45,10]
explode=(0.1,0.1,0.1,0.4)
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=label,autopct='%1.1f%%',shadow=True,startangle=90)
ax.axis("equal")
ax.set_title("Engineering discipline")
```
